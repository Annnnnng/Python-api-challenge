```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import gmaps.datasets
import json
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
file = "../WeatherPy/export_weather_data.csv"
df = pd.read_csv(file)
df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = df[["Lat", "Lon"]]
weights = df["Humidity"]
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(locations, weights=weights))
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
city_df = df.loc[(df["Max Temp"] < 280) \
                                    & (df["Wind Speed"] < 10) \
                                    & (df["Cloudiness"] == 0)].dropna()
hotel_df = city_df[["City", "Country", "Lat", "Lon"]].copy()
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Add Hotel Name column to dataframe
hotel_df["Hotel Name"] = ""
hotel_df
```

```python
# Get hotel information with API call
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
params = {"types": "lodging", "radius": 5000, "key": g_key}
hotel_list = []

for index, row in hotel_df.iterrows():
    
    lat = row["Lat"]
    lon = row["Lon"]
    params["location"] = f"{lat},{lon}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
    response = requests.get(url , params=params).json()
    hotel_list.append(response["results"])
pprint(hotel_list[0])
```

```python
# Loop through the api call response to get hotel name for each city
params = {"types": "lodging", "radius": 5000, "key": g_key}


for index, row in hotel_df.iterrows():
    
    lat = row["Lat"]
    lon = row["Lon"]
    params["location"] = f"{lat},{lon}"
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

    response = requests.get(url, params=params).json()
   

    try:
        hotel_df.loc[index, "Hotel Name"] = response["results"][0]["name"]
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```

```python
hotel_df.to_csv("Hotel.csv")
hotel_df
```
